# K-Means

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.cluster as cluster
import sklearn.datasets as datasets
```

```python
X, Y = datasets.load_iris(return_X_y=True)
N = len(Y)
```

```python
print(X.shape, Y.shape)
```

```python
plt.figure(figsize=(12,10))
for j in range(4):
    for i in range(j, 4):
        if i == j:
            continue
        k = 4*j+i
        plt.subplot(4, 4, k+1)
        plt.scatter(X[:,i], X[:,j], c=Y, s=16)
        plt.ylabel(str(j))
        plt.xlabel(str(i))
plt.tight_layout()
plt.show()
```

```python
km_1 = cluster.KMeans(2)
km_1.fit(X)
Y_p1 = km_1.predict(X)
c_1 = km_1.cluster_centers_
print(c_1)
```

```python
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.scatter(X[:,0], X[:,2], c=Y)
plt.subplot(1,2,2)
plt.scatter(X[:,0], X[:,2], c=Y_p1)
plt.scatter(c_1[:,0], c_1[:,2], s=64, c='r')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(X[:,1], X[:,2], c=Y)
plt.subplot(1,2,2)
plt.scatter(X[:,1], X[:,2], c=Y_p1)
plt.scatter(c_1[:,1], c_1[:,2], s=64, c='r')
plt.tight_layout()
plt.show()
```

```python
km_2 = cluster.KMeans(3)
km_2.fit(X)
Y_p2 = km_2.predict(X)
c_2 = km_2.cluster_centers_
print(c_2)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(X[:,1], X[:,2], c=Y)
plt.subplot(1,2,2)
plt.scatter(X[:,1], X[:,2], c=Y_p2)
plt.scatter(c_2[:,1], c_2[:,2], s=64, c='r')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(X[:,0], X[:,2], c=Y)
plt.subplot(1,2,2)
plt.scatter(X[:,0], X[:,2], c=Y_p2)
plt.scatter(c_1[:,0], c_1[:,2], s=36, c='r')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(X[:,1], X[:,2], c=Y)
for k in range(N):
    plt.text(X[k,1], X[k,2], str(Y[k]), color='k')
plt.subplot(1,2,2)
plt.scatter(X[:,1], X[:,2], c=Y_p2)
plt.scatter(c_2[:,1], c_2[:,2], s=36, c='r')
for k in range(N):
    plt.text(X[k,1], X[k,2], str(Y_p2[k]), color='k')
plt.tight_layout()
plt.show()
```

```python
Y_temp = Y_p2.copy()
Y_temp[Y_p2 == 1] = 0
Y_temp[Y_p2 == 0] = 1
# Y_temp[Y_p2 == 2] = 1
print(Y_p2)
print(Y_temp)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(X[:,1], X[:,2], c=Y)
for k in range(N):
    plt.text(X[k,1], X[k,2], str(Y[k]), color='k')
plt.subplot(1,2,2)
plt.scatter(X[:,1], X[:,2], c=Y_temp)
plt.scatter(c_2[:,1], c_2[:,2], s=36, c='r')
for k in range(N):
    plt.text(X[k,1], X[k,2], str(Y_temp[k]), color='k')
plt.tight_layout()
plt.show()
```

```python
import sklearn.metrics as metrics
```

```python
print(metrics.accuracy_score(Y, Y_temp))
```
