Подключаем библиотеки

```python
from tensorflow import keras
from google.colab import files
from PIL import Image
import numpy as np
from io import BytesIO
```

Выбираем сеть

```python
model = keras.applications.VGG16()
```

Чтобы изменить содержимое ячейки, дважды нажмите на нее (или выберите "Ввод")

```python
upl = files.upload()
names = list(upl.keys())
img = Image.open(BytesIO(upl[names[0]]))
img = img.resize((224, 224))
```

Приведение к входному формату VGG сети

```python
img = np.array(img)
x = keras.applications.vgg16.preprocess_input(img)
print(x.shape)
```

```python
x = np.expand_dims(x, axis=0)
print(x.shape)
```

Прогоняем через сеть

```python
res = model.predict(x)
print(np.argmax(res))
keras.applications.imagenet_utils.decode_predictions(res, top=5)
```
